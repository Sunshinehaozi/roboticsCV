#### Reading in Color Images
事实上感知应该分为多步进行，但都始于“reading in images”。为了实现自主导航，需要使用相机获取的图像，已决定可能形式的位置。通常这是一个十分复杂的问题，但在这里我们适当地简化处理。    
Rover所在环境中，地上的沙子十分明亮而其他地方比较灰暗。因此，判断Rover能够行驶的地方其实就是找出图像中明亮的地方。    
这里使用 matplotlib库(https://matplotlib.org/) 读取图像：    
(注：plt.imshow()负责处理图像，并显示格式，而plt.show()则是将plt.imshow()处理后的函数显示出来。)

```python
# Import some packages from matplotlib
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
# Uncomment the next line for use in a Jupyter notebook
%matplotlib inline

# Define the filename, read and plot the image
filename = 'sample.jpg'
image = mpimg.imread(filename)
plt.imshow(image)
plt.show
```

现在，图像已经存储为一个数组，你可以对它实施操作。首先，可以使用 numpy package (http://www.numpy.org/; Numpy应用于图像:http://scikit-image.org/docs/dev/user_guide/numpy_images.html) 查看它的大小，数据类型以及最大、最小值：

```python
# Import the "numpy" package for working with arrays
import numpy as np
print(image.dtype, image.shape, np.min(image), np.max(image))
#image
```

图像为一个元素为八位无符号整型的数组（8-bit unsigned integer array），它的shape为(160, 320, 3),这意味着图像的大小为：160 pixels in the y-direction (height), 320 pixels in the x-direction (width) and it has 3 layers or "color channels"。注意y轴方向为高度方向；x轴方向为宽度方向。     
图像的最大、最小值分别为0、255.这是因为图像的每个通道上存储有8位的信息（256个可能值，即0-255）。事实上并非所有图像都是按照这个格式存储的，所以在获取图像后，检查数据的类型是很有必要的。    
三个颜色通道分别为R,G,B。三个通道上的强度值组合决定了我们看到的图像颜色。你可以通过把其他的通道值归零，并仍旧将所有通道排列在一起，使用原来读取图像的方式，就可以看到每一个颜色通道。

```python
# Note: we use the np.copy() function rather than just saying red_channel = image
# because in Python, such a statement would set those two arrays equal to each other
# forever, meaning any changes made to one would also be made to the other!
red_channel = np.copy(image)
green_channel = np.copy(image)
blue_channel = np.copy(image)
# Note: here instead of extracting individual channels from the image
# I'll keep all 3 color channels in each case but set the ones I'm not interested 
# in to zero.
red_channel[:,:,[1, 2]] = 0 # Zero out the green and blue channels (or red_channel[:,:,1:])
green_channel[:,:,[0, 2]] = 0 # Zero out the red and blue channels
blue_channel[:,:,[0, 1]] = 0 # Zero out the red and green channels

fig = plt.figure(figsize=(12,3)) # Create a figure for plotting
plt.subplot(131) # Initialize subplot number 1 in a figure that is 3 columns 1 row
plt.imshow(red_channel) # Plot the red channel
plt.subplot(132) # Initialize subplot number 2 in a figure that is 3 columns 1 row
plt.imshow(green_channel)  # Plot the green channel
plt.subplot(133) # Initialize subplot number 3 in a figure that is 3 columns 1 row
plt.imshow(blue_channel)  # Plot the blue channel
plt.show()
```

#### Color Thresholding
正如我们所看到的，三个通道中，地面和天空亮度较大，而山体则比较灰暗。因此使用一个简单的颜色阈值是可能识别出地面像素的。接下来的任务是写一个函数，它输入一幅RGB图像和一个由三个通道的阈值组成的tuple，然后输出一个单通道二值图（像素值非0即1）。    
下面对比了

```python
# The bellow method uses Only R channel.
# Define a function to perform a color threshold
def color_thresh_1channel(img, rgb_thresh=(0, 0, 0)):
    # Create an array of zeros same xy size as img, but single channel
    color_select = np.zeros_like(img[:,:,0])
    # Apply the thresholds for R and assign 1's where threshold was exceeded
    # Return the single-channel binary image
    img_red = img[:,:,0]
    r = 0
    c = 0
    for pixel_row in img_red:  # Or use index to iterate: for i in range(img_red.shape[0])
        for pixel in pixel_row:
            if pixel > rgb_thresh[0]:
                color_select[r, c] = 1
            c += 1
        r += 1
        c = 0
    
    return color_select


# The bellow method uses 3 channel.
def color_thresh_3channel(img, rgb_thresh=(0, 0, 0)):
    # Create an array of zeros same xy size as img, but single channel
    color_select = np.zeros_like(img[:,:,0])
    # Require that each pixel be above all three threshold values in RGB
    # above_thresh will now contain a boolean array with "True"
    # where threshold was met
    above_thresh = (img[:,:,0] > rgb_thresh[0]) \
                & (img[:,:,1] > rgb_thresh[1]) \
                & (img[:,:,2] > rgb_thresh[2])
    # Index the array of zeros with the boolean array and set to 1
    color_select[above_thresh] = 1
    # Return the binary image
    return color_select
```

```python
# Define color selection criteria
red_threshold = 160
green_threshold = 160
blue_threshold = 160

rgb_threshold = (red_threshold, green_threshold, blue_threshold)
```

```python
# pixels below the thresholds
colorsel1 = color_thresh_1channel(image, rgb_thresh=rgb_threshold)
colorsel3 = color_thresh_3channel(image, rgb_thresh=rgb_threshold)
```

```python
# Display the original image and binary               
f, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(12, 3), sharey=True)
f.tight_layout()
ax1.imshow(image)
ax1.set_title('Original Image', fontsize=20)

ax2.imshow(colorsel1, cmap='gray')
ax2.set_title('1 channel Result', fontsize=20)
plt.subplots_adjust(left=0., right=1, top=0.9, bottom=0.)

ax3.imshow(colorsel3, cmap='gray')
ax3.set_title('3 channel Result', fontsize=20)
plt.subplots_adjust(left=0., right=1, top=0.9, bottom=0.)
plt.show() 
```
