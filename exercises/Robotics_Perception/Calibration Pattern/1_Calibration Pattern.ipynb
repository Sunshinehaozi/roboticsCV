### Calibration Pattern
校准相机意味着你必须将实际世界中的三维点(3D points in world)映射到相机中的二维图像平面(2D image plane in the camera)。在现实中，这是一个相当复杂的过程，但在接下来的练习中，我们会使用一些强大的opencv工具使相机校准变得容易。

第一步是选择一种标定靶标(calibration pattern)，在以下练习中，我们将使用棋盘图案(chessboard pattern)：
![chessboard pattern](chessboard pattern.png)

棋盘不是测试模式的唯一选择，也可以是由一些简单形状如圆构成的图案。但是无论使用何种形状，使用测试靶标(test pattern)的目的是提供一个你能够使用相机拍摄的已知几何图形，并在将世界中的3D点映射到2D图像时用作参考。

### Take Pictures of the Test Pattern

### Find Corners
在获得测试靶标的图像后，首先要做的就是在每幅图像中对测试靶标进行定位。    
在这次的练习中，你将使用OpenCV函数findChessboardCorners()和drawChessboardCorners()来自动寻找并在这些象棋靶标的图像中画出"inner corners"。  
将这两个函数应用于示例图像，你将可以得到这样的结果：    
![find corners](find corners.png)    

在下面的练习中，任务很简单。算出任何一行中的inside corners数量，并在变量nx中输入该值。类似地，算出给定列中的inside corners数量，并将其存储在nx中。记住，inside corners只是两个黑色和两个白色正方形相交的点，换句话说只对inside corners计数，而不包括outside corners。    

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import glob

# prepare object points
nx = 8#TODO: enter the number of inside corners in x
ny = 6#TODO: enter the number of inside corners in y

## Make a list of calibration images
#images = glob.glob('./Cal*.jpg')
## Select any index to grab an image from the list
#idx = -1
# Read in the image
#img = mpimg.imread(images[idx])
img = mpimg.imread('Calibration034.jpg')

# Convert to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)

# Find the chessboard corners
ret, corners = cv2.findChessboardCorners(gray, (nx, ny), None)

# If found, draw corners
if ret == True:
    # Draw and display the corners
    cv2.drawChessboardCorners(img, (nx, ny), corners, ret)
    plt.imshow(img)

print(ret)
```
